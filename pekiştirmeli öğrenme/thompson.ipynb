```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
veriler = pd.read_csv('Ads_CTR_Optimisation.csv')
```

```python
#Thompson
import random

N = 10000
d = 10
toplam = 0
secilenler = []
tiklamalar = [0]*d
oduller = [0]*d

for n in range(0,N):
    ad = 0
    max_th = 0
    for i in range(0,d):
        rasbeta = random.betavariate(oduller[i]+1,tiklamalar[i]-oduller[i]+1)
        if rasbeta > max_th:
            max_th = rasbeta
            ad = i
    secilenler.append(ad)
    tiklama = veriler.values[n,ad]
    tiklamalar[ad] = tiklamalar[ad] + 1
    oduller[ad] = oduller[ad] + tiklama
    toplam = toplam + tiklama

print(toplam)
```
